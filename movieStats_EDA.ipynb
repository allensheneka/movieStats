## Concatenate final_tmbd_*.csv.gz files

```python
# concatenate year 2000 and 2002 final_tmdb_data files along ROWS or vertical (since column names are the same)
# Modify code: https://www.geeksforgeeks.org/how-to-concatenate-two-or-more-pandas-dataframes/
final_tmdb_concat = pd.concat([final_tmdb_2000, final_tmdb_2001], axis=0)
display(final_tmdb_concat)
```

#### Average Budget and Revenue values seem low because movies with $0 values pull the average down

## Exploratory Data Analysis

```python
# How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
# Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.
tmdb_revenue = final_tmdb_concat[(final_tmdb_concat['budget'] > 0) & (final_tmdb_concat['revenue'] > 0)]
tmdb_revenue.info()
```

#### At least 375 movies had valid financial information

```python
# How many movies are there in each of the certification categories (G/PG/PG-13/R)?
# Note to self:  remember value_counts() only works on series w/ groupby not dataframe; use size()
tmdb_revenue.groupby(['certification']).size()
```

```python
# What is the average revenue per certification category?
tmdb_revenue.groupby(['certification']).revenue.mean()
```

```python
# format / suppress scientific notation
# Modified code: https://stackoverflow.com/questions/21137150/format-suppress-scientific-notation-from-pandas-aggregation-results
# reset to default by pd.reset_option('display.float_format')

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# revenue value without scientific notation
tmdb_revenue.groupby(['certification']).revenue.mean()
```

```python
# What is the average budget per certification category?
tmdb_revenue.groupby(['certification']).budget.mean()
```
